```python
import pandas as pd
import matplotlib.pyplot as plt

# Copy biar aman
df_plot = df.copy()

# Ubah ke datetime
df_plot['published_date'] = pd.to_datetime(df_plot['published_date'], errors='coerce', format='mixed')

# Ambil tahun
df_plot['tahun'] = df_plot['published_date'].dt.year
```

```python
tahun_counts = df_plot['tahun'].value_counts().sort_index()

plt.figure()
plt.plot(tahun_counts.index, tahun_counts.values, marker='o')
plt.title('Persebaran Artikel per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.grid()
plt.show()
```

```python
tag_counts = df_plot['Tag'].value_counts()

plt.figure()
plt.bar(tag_counts.index, tag_counts.values)
plt.xticks(rotation=45)
plt.title('Persebaran Tag')
plt.xlabel('Tag')
plt.ylabel('Jumlah')
plt.show()
```

```python
sentiment_counts = df_plot['Sentiment'].value_counts()

plt.figure()
plt.bar(sentiment_counts.index, sentiment_counts.values)
plt.title('Persebaran Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Jumlah')
plt.show()
```

```python
source_counts = df_plot['source'].value_counts().head(10)  # ambil top 10 biar ga terlalu padat

plt.figure()
plt.bar(source_counts.index, source_counts.values)
plt.xticks(rotation=45)
plt.title('Top 10 Source')
plt.xlabel('Source')
plt.ylabel('Jumlah')
plt.show()
```

```python
import pandas as pd
import re
from collections import Counter
import matplotlib.pyplot as plt
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

# Inisialisasi stopword Indonesia
factory = StopWordRemoverFactory()
stopwords_indo = set(factory.get_stop_words())

# Nama kolom teks dari df kamu
text_column = 'content'

# Gabungkan semua teks
all_text = ' '.join(df[text_column].dropna().astype(str))

# Cleaning
all_text = all_text.lower()
all_text = re.sub(r'[^a-zA-Z\s]', ' ', all_text)

# Tokenisasi
tokens = all_text.split()

# Hapus stopwords
filtered_tokens = [word for word in tokens if word not in stopwords_indo]

# Hitung frekuensi kata
word_counts = Counter(filtered_tokens)

# Ambil top 10 kata
top_words = word_counts.most_common(10)

# Pisahkan kata & jumlah
words = [word for word, count in top_words]
counts = [count for word, count in top_words]

# Visualisasi
plt.figure()
plt.bar(words, counts)
plt.xticks(rotation=45)
plt.title("Top 10 Kata dari Content (Bahasa Indonesia)")
plt.xlabel("Kata")
plt.ylabel("Frekuensi")
plt.show()
```
