# **Top TF-IDF Words**

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer

# Define the output folder
output_folder = "/content/drive/MyDrive/PBA_Project_A/image/"
os.makedirs(output_folder, exist_ok=True)

# Define the output file path
output_path = os.path.join(output_folder, "top_30_tfidf_terms.png")

# Define the text column
text_column = 'final_text'

# Prepare text data
def prepare_text(text):
    if isinstance(text, list):
        return ' '.join(text)
    return str(text)

texts = df[text_column].dropna().apply(prepare_text)

# Initialize TF-IDF vectorizer
vectorizer = TfidfVectorizer(
    max_df=0.8,
    min_df=1,
    ngram_range=(1, 2)
)

# Fit and transform the text data
tfidf_matrix = vectorizer.fit_transform(texts)

# Get feature names
feature_names = vectorizer.get_feature_names_out()

# Sum TF-IDF scores for each term
tfidf_scores = tfidf_matrix.sum(axis=0).A1

# Create a DataFrame for TF-IDF scores
tfidf_df = pd.DataFrame({
    'Term': feature_names,
    'TF-IDF Score': tfidf_scores
})

# Get the top 30 most important terms
top_n = 30
top_tfidf_df = tfidf_df.sort_values(
    by='TF-IDF Score',
    ascending=False
).head(top_n)

# Reverse order so the highest score appears at the top
top_tfidf_df = top_tfidf_df.iloc[::-1]

# Create the horizontal bar chart
plt.figure(figsize=(10, 12))

colors = sns.color_palette("viridis", len(top_tfidf_df))

plt.barh(
    top_tfidf_df['Term'],
    top_tfidf_df['TF-IDF Score'],
    color=colors,
    edgecolor='black',
    linewidth=0.5
)

# Add chart title and axis labels
plt.title(
    "Top 30 Most Important Terms (TF-IDF)",
    fontsize=18,
    fontweight='bold',
    pad=20
)

plt.xlabel("TF-IDF Score", fontsize=14)
plt.ylabel("Terms", fontsize=14)

# Add gridlines on the x-axis
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Adjust layout
plt.tight_layout()

# Save the plot
plt.savefig(output_path, dpi=300, bbox_inches='tight')

# Display the plot
plt.show()

print(f"Plot saved to: {output_path}")
```

# **TF-IDF Words for Each Group**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def top_tfidf_words(df, group_col, text_col='final_text', top_n=20):
    """
    Calculate the top TF-IDF words for each group.
    """

    results = {}

    for group in df[group_col].dropna().unique():
        subset = df[df[group_col] == group].copy()

        # Convert token lists into strings if needed
        texts = subset[text_col].dropna().apply(
            lambda x: ' '.join(x) if isinstance(x, list) else str(x)
        )

        # Remove empty text values
        texts = texts[texts.str.strip() != '']

        # Skip empty groups
        if texts.empty:
            continue

        # Avoid max_df error for very small groups
        if len(texts) > 1:
            vectorizer = TfidfVectorizer(max_df=0.8)
        else:
            vectorizer = TfidfVectorizer()

        tfidf_matrix = vectorizer.fit_transform(texts)
        feature_names = vectorizer.get_feature_names_out()
        mean_scores = tfidf_matrix.mean(axis=0).A1
        top_indices = mean_scores.argsort()[::-1][:top_n]

        top_words = [(feature_names[i], mean_scores[i]) for i in top_indices]
        results[str(group)] = top_words

    return results
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the output folder
output_folder = "/content/drive/MyDrive/PBA_Project_A/image/"
os.makedirs(output_folder, exist_ok=True)

def plot_tfidf_by_group(top_words_dict, group_name, title_prefix, color_palette='viridis', top_n=20):
    """
    Create and save a small horizontal bar chart for one group.

    Args:
        top_words_dict (dict): Dictionary of TF-IDF results.
        group_name (str): Group name to visualize.
        title_prefix (str): Title prefix, e.g. 'Sentiment' or 'Quarter'.
        color_palette (str): Seaborn color palette.
        top_n (int): Number of top words to plot.
    """

    # Check whether the group exists
    if group_name not in top_words_dict:
        print(f"Group '{group_name}' not found.")
        print("Available groups:", list(top_words_dict.keys()))
        return

    # Create a dataframe
    plot_df = pd.DataFrame(
        top_words_dict[group_name],
        columns=['Term', 'TF-IDF Score']
    ).head(top_n)

    # Reverse order so the highest value is shown at the top
    plot_df = plot_df.iloc[::-1]

    # Create the chart
    plt.figure(figsize=(6, 4.5))
    colors = sns.color_palette(color_palette, len(plot_df))

    plt.barh(
        plot_df['Term'],
        plot_df['TF-IDF Score'],
        color=colors,
        edgecolor='white'
    )

    # Add title and labels
    plt.title(
        f"Top {top_n} TF-IDF Terms - {title_prefix} {group_name}",
        fontsize=12,
        fontweight='bold',
        pad=10
    )
    plt.xlabel("TF-IDF Score", fontsize=10)
    plt.ylabel("Terms", fontsize=10)

    # Add gridlines
    plt.grid(axis='x', linestyle='--', alpha=0.5)

    # Adjust layout
    plt.tight_layout()

    # Save the plot
    safe_group_name = str(group_name).replace(" ", "_")
    output_path = os.path.join(
        output_folder,
        f"tfidf_{title_prefix.lower()}_{safe_group_name}.png"
    )
    plt.savefig(output_path, dpi=300, bbox_inches='tight')

    # Show the plot
    plt.show()

    print(f"Plot saved to: {output_path}")
```

```python
import stanza
import pandas as pd
import ast

# Download Indonesian model
stanza.download('id')

# Load Indonesian NLP pipeline for POS tagging only
nlp_pos = stanza.Pipeline(
    lang='id',
    processors='tokenize,pos'
)
```

```python
def prepare_text_for_nlp(text):
    """
    Convert final_text into a normal string.
    Handles list, string-list, and normal string formats.
    """

    if isinstance(text, list):
        return ' '.join(text)

    if isinstance(text, str):
        try:
            parsed = ast.literal_eval(text)
            if isinstance(parsed, list):
                return ' '.join(parsed)
        except:
            pass

        return text

    return ""
```

```python
def get_pos_tags(text):
    """
    Perform POS tagging using Stanza.
    """

    text = prepare_text_for_nlp(text)

    if not text.strip():
        return []

    doc = nlp_pos(text)

    pos_tags = []

    for sentence in doc.sentences:
        for word in sentence.words:
            pos_tags.append({
                'word': word.text,
                'upos': word.upos,
                'xpos': word.xpos
            })

    return pos_tags
```

```python
# Apply POS tagging to final_text
df['pos_tags'] = df['final_text'].apply(get_pos_tags)

# Preview the result
df[['final_text', 'pos_tags']].head()
```

# **Named Entity Recognition**

```python
import re
import pandas as pd

# Define BPJS PBI-related entity patterns
entity_patterns = {
    'ORGANIZATION': [
        'BPJS',
        'BPJS Kesehatan',
        'Kementerian Kesehatan',
        'Dinas Sosial',
        'Pemerintah',
        'Pemerintah Daerah',
        'Pemda'
    ],
    'PROGRAM': [
        'PBI',
        'JKN',
        'KIS',
        'Penerima Bantuan Iuran',
        'Jaminan Kesehatan Nasional',
        'Kartu Indonesia Sehat'
    ],
    'HEALTH_FACILITY': [
        'Puskesmas',
        'Rumah Sakit',
        'Faskes',
        'Klinik'
    ]
}
```

```python
def get_rule_based_entities(text):
    """
    Extract named entities using rule-based matching for BPJS PBI-related terms.
    """

    text = prepare_text_for_nlp(text)

    entities = []

    for label, terms in entity_patterns.items():
        for term in terms:
            pattern = re.compile(r'\b' + re.escape(term) + r'\b', re.IGNORECASE)

            for match in pattern.finditer(text):
                entities.append({
                    'entity': match.group(),
                    'label': label
                })

    return entities
```

```python
# Apply rule-based NER to final_text
df['named_entities'] = df['final_text'].apply(get_rule_based_entities)

# Preview the result
df[['final_text', 'named_entities']].head()
```

```python
import re
import html
from IPython.display import display, HTML

# Define colors for each entity label
entity_colors = {
    'ORGANIZATION': '#cfe8ff',
    'PROGRAM': '#d9f7be',
    'HEALTH_FACILITY': '#ffd6d6'
}

def get_rule_based_entities_with_span(text):
    """
    Extract named entities with start and end positions.
    """

    text = str(text)
    entities = []

    for label, terms in entity_patterns.items():
        for term in terms:
            pattern = re.compile(r'\b' + re.escape(term) + r'\b', re.IGNORECASE)

            for match in pattern.finditer(text):
                entities.append({
                    'entity': match.group(),
                    'label': label,
                    'start': match.start(),
                    'end': match.end()
                })

    # Sort by start position, then by longest entity first
    entities = sorted(entities, key=lambda x: (x['start'], -(x['end'] - x['start'])))

    # Remove overlapping entities
    filtered_entities = []
    last_end = -1

    for ent in entities:
        if ent['start'] >= last_end:
            filtered_entities.append(ent)
            last_end = ent['end']

    return filtered_entities
```

```python
def highlight_entities_safe(text, entities):
    """
    Highlight entities safely without repeated labels.
    """

    text = str(text)
    entities = sorted(entities, key=lambda x: x['start'])

    html_parts = []
    last_idx = 0

    for ent in entities:
        start = ent['start']
        end = ent['end']
        entity = text[start:end]
        label = ent['label']
        color = entity_colors.get(label, '#eeeeee')

        # Add text before entity
        html_parts.append(html.escape(text[last_idx:start]))

        # Add highlighted entity
        html_parts.append(
            f"<mark style='background-color:{color}; padding:2px 5px; border-radius:5px;'>"
            f"{html.escape(entity)} "
            f"<span style='font-size:10px; font-weight:bold;'>[{label}]</span>"
            f"</mark>"
        )

        last_idx = end

    # Add remaining text after the last entity
    html_parts.append(html.escape(text[last_idx:]))

    return ''.join(html_parts)
```

```python
sample_index = 0

sample_text = df.loc[sample_index, 'content']  # better for NER visualization
sample_entities = get_rule_based_entities_with_span(sample_text)

html_text = highlight_entities_safe(sample_text, sample_entities)

display(HTML(f"""
<h3>NER Visualization</h3>
<p style="line-height:1.8; font-size:15px;">
{html_text}
</p>
"""))
```
