# **Show Sentiment Distribution**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Count the number of data for each sentiment
sentiment_counts = df['Sentiment'].value_counts()

# Calculate the percentage of each sentiment
sentiment_percentage = df['Sentiment'].value_counts(normalize=True) * 100

# Combine counts and percentages into one DataFrame
sentiment_summary = pd.DataFrame({
    'Count': sentiment_counts,
    'Percentage': sentiment_percentage.round(2)
})

# Display sentiment distribution
print("Sentiment Distribution:")
print(sentiment_summary)

# Visualize sentiment distribution
plt.figure(figsize=(8, 5))

bars = plt.bar(
    sentiment_summary.index,
    sentiment_summary['Count'],
    edgecolor='black',
    linewidth=0.5,
    alpha=0.9
)

# Add chart title and axis labels
plt.title(
    "Sentiment Distribution",
    fontsize=16,
    fontweight='bold',
    pad=15
)

plt.xlabel("Sentiment", fontsize=12)
plt.ylabel("Count", fontsize=12)

# Add value labels on top of each bar
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        int(bar.get_height()),
        ha='center',
        va='bottom',
        fontsize=11
    )

# Add gridlines on the y-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Check class imbalance ratio
max_count = sentiment_counts.max()
min_count = sentiment_counts.min()
imbalance_ratio = max_count / min_count

print(f"\nImbalance ratio: {imbalance_ratio:.2f}")

if imbalance_ratio >= 2:
    print("The dataset is imbalanced. Balancing is recommended.")
else:
    print("The dataset is relatively balanced. Balancing may not be necessary.")
```

# **Define augmented function**

```python
import random
import re

# Define protected words that should not be replaced
protected_words = {
    'bpjs', 'pbi', 'jkn', 'kis',
    'kesehatan', 'iuran',
    'peserta', 'penerima', 'bantuan',
    'jaminan', 'layanan',
    'faskes', 'puskesmas', 'rumah', 'sakit',
    'kelas', 'aktif', 'nonaktif'
}

# Define custom Indonesian synonyms for BPJS PBI-related text
bpjs_pbi_synonyms_dict = {
    # General verbs
    "membantu": ["menolong", "mendukung"],
    "memberikan": ["menyediakan", "menyalurkan"],
    "menerima": ["mendapat", "memperoleh"],
    "mendapatkan": ["memperoleh", "menerima"],
    "menggunakan": ["memakai", "memanfaatkan"],
    "mengakses": ["mendapatkan akses ke", "memanfaatkan"],
    "menanggung": ["membiayai", "menjamin"],
    "membayar": ["melunasi", "menyetor"],
    "mengurus": ["memproses", "menangani"],
    "mendaftarkan": ["meregistrasikan", "mencatatkan"],
    "terdaftar": ["tercatat", "teregistrasi"],
    "menonaktifkan": ["membekukan", "menghentikan"],
    "mengaktifkan": ["menghidupkan kembali", "mengaktifkan kembali"],
    "memeriksa": ["mengecek", "meninjau"],
    "mengecek": ["memeriksa", "melihat"],
    "meningkatkan": ["menaikkan", "memperbaiki"],
    "menurunkan": ["mengurangi", "menekan"],

    # BPJS/PBI-related nouns
    "masyarakat": ["warga", "publik"],
    "warga": ["masyarakat", "penduduk"],
    "penduduk": ["warga", "masyarakat"],
    "pemerintah": ["negara", "otoritas"],
    "daerah": ["wilayah", "kawasan"],
    "data": ["informasi", "catatan"],
    "program": ["skema", "kebijakan"],
    "kebijakan": ["aturan", "ketentuan"],
    "aturan": ["ketentuan", "regulasi"],
    "bantuan": ["dukungan", "subsidi"],
    "subsidi": ["bantuan", "dukungan"],
    "pelayanan": ["layanan", "servis"],
    "akses": ["jangkauan", "kesempatan"],
    "fasilitas": ["sarana", "prasarana"],
    "rumah_sakit": ["fasilitas_kesehatan"],
    "puskesmas": ["fasilitas_kesehatan"],
    "administrasi": ["pengurusan", "tata_usaha"],
    "verifikasi": ["pemeriksaan", "pengecekan"],
    "validasi": ["pengesahan", "pengecekan"],
    "keluhan": ["aduan", "komplain"],
    "masalah": ["kendala", "persoalan"],
    "kendala": ["hambatan", "masalah"],
    "biaya": ["ongkos", "pengeluaran"],
    "tagihan": ["billing", "tanggungan"],

    # Adjectives/adverbs
    "aktif": ["berlaku"],
    "nonaktif": ["tidak_aktif"],
    "gratis": ["tanpa_biaya"],
    "mudah": ["gampang", "praktis"],
    "sulit": ["susah", "rumit"],
    "penting": ["utama", "krusial"],
    "layak": ["memenuhi_syarat"],
    "tidak_layak": ["tidak_memenuhi_syarat"],
    "cepat": ["segera", "lekas"],
    "lambat": ["pelan", "terhambat"],
    "tepat": ["akurat", "sesuai"],
    "sesuai": ["tepat", "cocok"],
    "belum": ["masih_belum"],
    "sudah": ["telah"]
}
```

```python
import pandas as pd
import re
import random
import time
import logging
from multiprocessing import Pool
from retry import retry
from deep_translator import GoogleTranslator
from sklearn.model_selection import train_test_split

# ==============================
# 1. SETUP & CONFIG
# ==============================
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

TEXT_COL = 'removed_stopwords_content'
AUG_COL = 'augmented_content' # Kolom baru untuk hasil augmentasi
LABEL_COL = 'Sentiment'

# Dummy dict (Pastikan ini sudah didefinisikan di kodemu)
# bpjs_pbi_synonyms_dict = {...}
# protected_words = [...]

# ==============================
# 2. CLEANING FUNCTION
# ==============================
def clean_news_junk(text):
    if not text: return ""
    text = str(text)
    text = re.sub(r"[\[\]']", "", text)
    text = re.sub(r"\d{2} [a-zA-Z]+ \d{4}.*", "", text)
    text = re.sub(r"copyright.*all rights reserved", "", text, flags=re.IGNORECASE)
    return text[:2500]

# ==============================
# 3. AUGMENTATION WORKERS
# ==============================
def synonym_replacement(text, rate=0.15): # Rate dikecilkan agar tidak merusak makna
    words = text.split()
    if len(words) < 5: return text
    new_words = []
    for word in words:
        clean_w = re.sub(r'[^a-zA-Z]', '', word.lower())
        if clean_w in bpjs_pbi_synonyms_dict and random.random() < rate and clean_w not in protected_words:
            new_words.append(random.choice(bpjs_pbi_synonyms_dict[clean_w]))
        else:
            new_words.append(word)
    return " ".join(new_words)

@retry(tries=3, delay=5, backoff=2)
def back_translate(text):
    if len(text) < 20: return text # Hanya teks panjang yang masuk akal di-BT
    try:
        en_text = GoogleTranslator(source='id', target='en').translate(text)
        time.sleep(random.uniform(1.2, 2.5))
        return GoogleTranslator(source='en', target='id').translate(en_text)
    except Exception as e:
        logger.error(f"BT Error: {e}")
        return text

def augment_process(row):
    # Gunakan teks asli sebagai dasar augmentasi
    original_txt = clean_news_junk(row[TEXT_COL])

    # Logika: 70% Back-translation (karena lebih berkualitas untuk BERT)
    # 30% Hanya Synonym Replacement
    if random.random() > 0.3:
        bt_txt = back_translate(original_txt)
        # Tambahkan SR sedikit setelah BT untuk variasi ekstra
        row[AUG_COL] = synonym_replacement(bt_txt, rate=0.05)
        row['aug_method'] = 'Back-Translation + SR'
    else:
        row[AUG_COL] = synonym_replacement(original_txt, rate=0.2)
        row['aug_method'] = 'Synonym Replacement Only'

    row['is_augmented'] = True
    return row

# ==============================
# 4. RUNNER (TARGETED BALANCING)
# ==============================
if __name__ == '__main__':
    # 1. Pisahkan data berdasarkan label
    df_positif = df[df[LABEL_COL] == 'Positif'].copy()
    df_negatif = df[df[LABEL_COL] == 'Negatif'].copy()
    df_netral = df[df[LABEL_COL] == 'Netral'].copy()

    # 2. Ambil 46 sampel dari Netral untuk diaugmentasi (BT + SR)
    # n=46 agar Total Netral menjadi 174 + 46 = 220
    df_netral_to_aug = df_netral.sample(n=46, random_state=42)

    # 3. Proses Augmentasi Data Netral (BT + SR)
    logger.info("Memulai augmentasi 46 data Netral (BT + SR)...")
    with Pool(processes=2) as p:
        results_netral = p.map(augment_process, [row for _, row in df_netral_to_aug.iterrows()])
    df_aug_netral = pd.DataFrame(results_netral)

    # 4. Proses Kelas Lainnya (Hanya SR)
    # Sesuai permintaan: Positif dan Negatif diproses SR tanpa penerjemah
    logger.info("Memproses kelas Positif dan Negatif dengan Synonym Replacement saja...")

    # Gabungkan Positif dan Negatif untuk diproses sekaligus
    df_others = pd.concat([df_positif, df_negatif], ignore_index=True)

    with Pool(processes=2) as p:
        results_others = p.map(augment_process, [row for _, row in df_others.iterrows()])
    df_others_processed = pd.DataFrame(results_others)

    # 5. Gabungkan Semua Kembali
    # Urutan: Positif&Negatif(SR) + Netral(Asli) + Netral(BT+SR)
    df_final = pd.concat([df_others_processed, df_netral, df_aug_netral], ignore_index=True)

    # ==============================
    # 5. SAVE & VERIFIKASI
    # ==============================
    print("\n--- Distribusi Akhir Sesuai Target ---")
    print(df_final[LABEL_COL].value_counts())

    # Harusnya:
    # Positif: 220
    # Negatif: 211
    # Netral : 220 (174 asli + 46 augmented)

    df_final.to_csv('data_augmented_targeted.csv', index=False)
    logger.info("Data berhasil disimpan di 'data_augmented_targeted.csv'")
```
