# **Sentiment Label and Train Test Split**

```python
mapping = {
    'Negatif': 0,
    'Netral': 1,
    'Positif': 2
}

df['label'] = df['Sentiment'].map(mapping)
```

```python
import re
def clean_list_string(text):
    text = str(text)
    # Hapus kurung siku, kutip tunggal, dan kutip dua
    text = re.sub(r"[[\\]'", " ", text)
    # Hapus spasi ganda yang muncul akibat penghapusan karakter tadi
    text = re.sub(r'\\s+', ' ', text).strip()
    return text

# Terapkan ke kolom teksmu
df['augmented_content'] = df['augmented_content'].apply(clean_list_string)
```

```python
x_data = df['augmented_content'].astype(str)
y_data = df['label']

xtrain, xtest, ytrain, ytest = train_test_split(
    x_data, y_data,
    test_size=0.2,
    random_state=42,
    stratify=y_data
)
```

```python
label_names = {
    0: 'Negatif',
    1: 'Netral',
    2: 'Positif'
}

print(f'Training set : {len(xtrain):,} sampel')
print(f'Testing set  : {len(xtest):,} sampel')
print()

for split_name, y_split in [('Train', ytrain), ('Test', ytest)]:
    print(f'{split_name} distribution:')
    counts = pd.Series(y_split).value_counts().sort_index()
    for idx, cnt in counts.items():
        print(f'  {label_names[idx]:<12}: {cnt:,} ({cnt/len(y_split)*100:.1f}%)')
    print()
```

```python
from collections import Counter

# Counting the frequency of each word in the training data
word_counts = Counter(' '.join(str(item) for item in xtrain).split())
print(f'Total kata unik: {len(word_counts):,}')
for word, count in word_counts.most_common(100):
    print(f"{word}: {count}")
```

# **Initialize Classifiers**

```python
def init_classifiers():
    svm_classifier = LinearSVC(
        max_iter=2000,
        random_state=42,
        class_weight='balanced'
    )

    logistic_regression_classifier = LogisticRegression(
        max_iter=1000,
        random_state=42,
        class_weight='balanced'
    )

    return svm_classifier, logistic_regression_classifier
```

```python
def evaluate_model(model, xtest_data, ytest_labels):
    y_pred = model.predict(xtest_data)

    accuracy = accuracy_score(ytest_labels, y_pred)
    precision = precision_score(ytest_labels, y_pred, average='weighted', zero_division=0)
    recall = recall_score(ytest_labels, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(ytest_labels, y_pred, average='weighted', zero_division=0)

    target_names = ['Negatif', 'Netral', 'Positif']

    report = classification_report(
        ytest_labels,
        y_pred,
        labels=[0,1,2],
        target_names=target_names,
        zero_division=0
    )

    cm = confusion_matrix(ytest_labels, y_pred, labels=[0,1,2])

    return y_pred, accuracy, precision, recall, f1, report, cm
```

# **TF-IDF Vectorization**

```python
from timeit import default_timer as timer

print('=' * 70)
print('TF-IDF VECTORIZATION')
print('=' * 70)

tfidf_vectorizer = TfidfVectorizer(
    max_features=10000,
    ngram_range=(1,2)
)

# Fit
start = timer()
tfidf_vectorizer.fit(xtrain)
print(f'Vectorizer fitted on training data in {timer() - start:.2f}s')
print()

# Transform train
start = timer()
xtrain_tfidf = tfidf_vectorizer.transform(xtrain)
train_time = timer() - start

# Transform test
start = timer()
xtest_tfidf = tfidf_vectorizer.transform(xtest)
test_time = timer() - start

print(f'TF-IDF training matrix shape : {xtrain_tfidf.shape}')
print(f'TF-IDF testing matrix shape  : {xtest_tfidf.shape}')
print(f'Matrix type                  : {type(xtrain_tfidf).__name__}')
print(f'Training transform time      : {train_time:.2f}s')
print(f'Testing transform time       : {test_time:.2f}s')
print()
```

```python
print('=' * 70)
print('MODEL TRAINING WITH TF-IDF FEATURES')
print('=' * 70)

# Initialize the classifiers
svm_model, lr_model = init_classifiers()

print('Training all classifiers on TF-IDF features...')
print()

start = timer()
svm_tfidf = svm_model.fit(xtrain_tfidf, ytrain)
print(f'Linear SVM          : {timer() - start:.2f}s')

start = timer()
lr_tfidf = lr_model.fit(xtrain_tfidf, ytrain)
print(f'Logistic Regression : {timer() - start:.2f}s')

print()
print('TF-IDF-based model training has been completed.')
```

## **Linear SVM TF-IDF**

```python
print('=' * 70)
print('LINEAR SVM PERFORMANCE ON TF-IDF FEATURES')
print('=' * 70)

y_pred_svm, acc_svm, prec_svm, rec_svm, f1_svm, rep_svm, cm_svm = evaluate_model(
    svm_tfidf, xtest_tfidf, ytest
)

print('\nSample Predictions')
print('-' * 70)
for i, (actual, predicted) in enumerate(zip(ytest[:15], y_pred_svm[:15]), start=1):
    mark = '✓' if actual == predicted else '✗'
    print(f'{i:>2}. {mark} Actual: {label_names[actual]:<12} Predicted: {label_names[predicted]}')

print('\nOverall Performance')
print('-' * 70)
print(f'Accuracy  : {acc_svm:.4f}')
print(f'Precision : {prec_svm:.4f}')
print(f'Recall    : {rec_svm:.4f}')
print(f'F1-Score  : {f1_svm:.4f}')

print('\nDetailed Classification Report')
print('-' * 70)
print(rep_svm)

print('Confusion Matrix')
print('-' * 70)
print(cm_svm)
```

## **IndoBERT Fine Tuning**

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

model_name = "indobenchmark/indobert-base-p1"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=3
)
```

```python
import re
def finalize_cleaning(text):
    text = str(text)
    # Hapus format list ['kata', 'kata'] jika ada
    text = re.sub(r"[[\]']", "", text)
    # Hapus spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()
    # Potong teks jika terlalu panjang (Ambil 200 kata pertama saja)
    words = text.split()
    return " ".join(words[:200])

# Terapkan ke data sebelum training
xtrain = xtrain.apply(finalize_cleaning)
xtest = xtest.apply(finalize_cleaning)
```

```python
def tokenize_function(texts):
    return tokenizer(
        texts.tolist(),
        padding=True,
        truncation=True,
        max_length=256
    )

train_encodings = tokenize_function(xtrain)
test_encodings = tokenize_function(xtest)
```

```python
import torch

class SentimentDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels.values

    def __getitem__(self, idx):
        item = {k: torch.tensor(v[idx]) for k, v in self.encodings.items()}
        item["labels"] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = SentimentDataset(train_encodings, ytrain)
test_dataset = SentimentDataset(test_encodings, ytest)
```

```python
from transformers import EarlyStoppingCallback, TrainingArguments, Trainer
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = logits.argmax(axis=1)

    precision, recall, f1, _ = precision_recall_fscore_support(
        labels, preds, average='weighted', zero_division=0
    )
    acc = accuracy_score(labels, preds)

    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall
    }

# 1. Update Argumen: Tambahkan fp16 dan naikkan batch size
training_args = TrainingArguments(
    output_dir="./results",
    learning_rate=3e-5,              # Sedikit lebih tinggi untuk eksplorasi
    per_device_train_batch_size=32,  # Naikkan jika VRAM GPU mencukupi
    per_device_eval_batch_size=32,
    num_train_epochs=15,             # Lebih banyak waktu belajar
    weight_decay=0.015,              # Regulasi lebih kuat agar tidak overfit
    warmup_ratio=0.1,                # 10% langkah awal untuk pemanasan
    lr_scheduler_type="cosine",      # Penurunan LR yang lebih halus
    eval_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    fp16=True,                       # Tetap gunakan untuk kecepatan GPU
    save_total_limit=1
)

# 2. Pastikan Callback terpasang hanya di SATU Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    compute_metrics=compute_metrics,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=4)]
)

# 3. Jalankan Training
trainer.train()
```

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = logits.argmax(axis=1)

    precision, recall, f1, _ = precision_recall_fscore_support(
        labels, preds, average='weighted', zero_division=0
    )
    acc = accuracy_score(labels, preds)

    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall
    }
```

```python
predictions = trainer.predict(test_dataset)
y_pred_bert = predictions.predictions.argmax(axis=1)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

label_names = {
    0: 'Negatif',
    1: 'Netral',
    2: 'Positif'
}

print('=' * 70)
print('BERT FINE-TUNING PERFORMANCE')
print('=' * 70)

print('\nSample Predictions')
print('-' * 70)
for i, (actual, predicted) in enumerate(zip(ytest[:15], y_pred_bert[:15]), start=1):
    mark = '✓' if actual == predicted else '✗'
    print(f'{i:>2}. {mark} Actual: {label_names[actual]:<12} Predicted: {label_names[predicted]}')

print('\nOverall Performance')
print('-' * 70)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

acc = accuracy_score(ytest, y_pred_bert)
prec = precision_score(ytest, y_pred_bert, average='weighted', zero_division=0)
rec = recall_score(ytest, y_pred_bert, average='weighted', zero_division=0)
f1 = f1_score(ytest, y_pred_bert, average='weighted', zero_division=0)

print(f'Accuracy  : {acc:.4f}')
print(f'Precision : {prec:.4f}')
print(f'Recall    : {rec:.4f}')
print(f'F1-Score  : {f1:.4f}')

print('\nDetailed Classification Report')
print('-' * 70)
print(classification_report(
    ytest,
    y_pred_bert,
    labels=[0,1,2],
    target_names=['negatif','netral','positif'],
    zero_division=0
))

print('Confusion Matrix')
print('-' * 70)
print(confusion_matrix(ytest, y_pred_bert, labels=[0,1,2]))
```
